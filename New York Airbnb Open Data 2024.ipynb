**Distribución de precios por tipo de alquiler**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Establecer el estilo y el tamaño de la figura
sns.set(style="whitegrid")
plt.figure(figsize=(10, 10))

# Definir la paleta de colores
colores = ['#007A87', '#FF5A5F', '#484848', '#00C897']

# Crear el gráfico de caja con la paleta de colores especificada
sns.boxplot(data=dataset, x='room_type', y='price', palette=colores)
plt.yscale('log')  # Escala logarítmica para el eje y

# Etiquetas y título
plt.xlabel('Tipo de Habitación', fontsize=14)
plt.ylabel('Precio (escala log)', fontsize=14)
plt.title('Distribución de precios por tipo de alquiler', fontsize=16, fontweight='bold', pad=20, loc='left')

plt.show()
```

**Proporción de propiedades por distrito**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Definir la paleta de colores
colores = ['#007A87', '#FF5A5F', '#484848', '#00C897', '#FFC400']

# Agrupar por distrito y contar el número de propiedades en cada uno
propiedades_por_barrio_grupo = dataset['neighbourhood_group'].value_counts()

# Obtener los valores y etiquetas para el gráfico de pie
valores = propiedades_por_barrio_grupo.values
etiquetas = propiedades_por_barrio_grupo.index

# Crear el gráfico pie
plt.figure(figsize=(8, 8))
plt.pie(valores, labels=etiquetas, autopct='%1.1f%%', startangle=140, colors=colores)

# Agregar centro hueco
centro_circulo = plt.Circle((0, 0), 0.70, color='white')
fig = plt.gcf()
fig.gca().add_artist(centro_circulo)

# Añadir título
plt.title('Proporción de propiedades por distrito', loc='center', fontsize=16, fontweight='bold', pad=20)

# Mostrar el gráfico
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()
plt.show()
```

**Número de propiedades en los 10 barrios con más disponibilidad**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Agrupar por barrio y contar el número de propiedades en cada barrio
propiedades_por_barrio = dataset['neighbourhood'].value_counts()

# Tomar los 10 barrios con más propiedades y ordenarlos de mayor a menor
top_10_barrios = propiedades_por_barrio.head(10).sort_values(ascending=False)

# Definir la paleta de colores
colores = ['#007A87', '#FF5A5F', '#484848', '#00C897', '#FFC400', '#FF3839', '#1BB270', '#A2A2A2', '#FFD043', '#00A699']

# Crear el gráfico de barras horizontales con Seaborn y aplicar la paleta de colores
plt.figure(figsize=(10, 6))
grafico = sns.barplot(x=top_10_barrios.values, y=top_10_barrios.index, palette=colores)

# Añadir los números fuera de las barras
for i, valor in enumerate(top_10_barrios.values):
    plt.text(valor + 0.1, i, str(valor), ha='left', va='center')

# Título y etiquetas
plt.title('Número de propiedades en los 10 barrios con más disponibilidad', fontsize=16, fontweight='bold', pad=20, loc='left')
plt.xlabel('Número de Propiedades', fontsize=14)
plt.ylabel('Barrio', fontsize=14)

plt.show()
```

**Cantidad de propiedades disponibles por tipo de alquiler**

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Definir paleta de colores
colores = ['#007A87', '#FF5A5F', '#484848', '#00C897', '#FFC400', '#FF3839', '#1BB270', '#A2A2A2', '#FFD043', '#00A699']

# Agrupando y contando por tipo de alquiler
room_type_counts = dataset['room_type'].value_counts().reset_index()
room_type_counts.columns = ['room_type', 'count']

# Definiendo el grafico de barras horizontal
plt.figure(figsize=(11, 6))
ax = sns.barplot(data=room_type_counts, x='count', y='room_type', palette=colores)

# Añadir los números fuera de las barras
for index, value in enumerate(room_type_counts['count']):
    ax.text(value, index, str(value), va='center', ha='left', fontsize=12, color='black')

# Titulos y etiquetas
plt.title('Cantidad de propiedades disponibles por tipo de alquiler', fontsize=16, fontweight='bold', pad=20, loc='left')
plt.xlabel('Cantidad', fontsize=14)
plt.ylabel('Tipo de habitación', fontsize=14)

# Mostrando el grafico
plt.tight_layout()
plt.show()
```

**Distribución de precios por distrito**

**Histograma de la duración mínima de la estadía**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Configurando el estilo y la paleta de colores
sns.set_style("whitegrid")
sns.set_palette("Set3")

# Creando el histograma de la duración mínima de la estadía
plt.figure(figsize=(10, 6))
sns.histplot(data=dataset, x='minimum_nights', bins=30, kde=True)  # Ajusta el número de bins según sea necesario

# Titulo y etiquetas
plt.title('Histograma de la duración mínima de la estadía', fontsize=16,fontweight='bold', pad=20, loc='left')
plt.xlabel('Duración mínima de la estadía (# de noches)', fontsize=14)
plt.ylabel('Frecuencia', fontsize=14)

# Limite para mostrar en el eje x
plt.xlim(0, 100)

# Calcular el valor máximo de la frecuencia
max_frequency = dataset['minimum_nights'].max()

# Encontrar el valor de duración mínima de la estadía correspondiente al punto máximo
max_value = dataset['minimum_nights'].mode().values[0]

# Anotar el valor de duración mínima de la estadía en el punto máximo del histograma
plt.text(max_value, max_frequency, f'Mas frecuente: {max_value}', ha='center', fontsize=12)

plt.tight_layout()
plt.show()
```

**Disponibilidad media de los alquileres a lo largo del año**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Convertir la columna 'last_review' a tipo datetime
dataset['last_review'] = pd.to_datetime(dataset['last_review'])

# Extraer el mes de la columna 'last_review'
dataset['month'] = dataset['last_review'].dt.month

# Calcular la disponibilidad media para cada mes
availability_by_month = dataset.groupby('month')['availability_365'].mean()

# Configurando el estilo y la paleta de colores
sns.set_style("whitegrid")
sns.set_palette("Set3")

# Crear el gráfico de líneas para mostrar la disponibilidad media a lo largo del año
plt.figure(figsize=(10, 6))
availability_by_month.plot(marker='o', color='tab:blue', linewidth=2, markersize=8)

# Titulo y etiquetas
plt.title('Disponibilidad media de los alquileres a lo largo del año', fontsize=16,fontweight='bold', pad=20, loc='left')
plt.xlabel('Mes', fontsize=14)
plt.ylabel('Disponibilidad media', fontsize=14)
plt.xticks(range(1, 13), ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic'])

plt.tight_layout()
plt.show()
```

**Matriz de correlación entre variables Numéricas**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Seleccionar solo las variables numéricas del DataFrame para calcular la correlación
numeric_variables = dataset.select_dtypes(include=['int64', 'float64'])

# Calcular la matriz de correlación
correlation_matrix = numeric_variables.corr()

# Configurar el tamaño del gráfico
plt.figure(figsize=(10, 8))

# Crear el mapa de calor
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Matriz de correlación entre variables Numéricas', fontsize=16,fontweight='bold', pad=20, loc='left')
plt.show()
```

**Distribución de reviews por mes**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Convertir la columna 'last_review' a tipo datetime si es necesario
dataset['last_review'] = pd.to_datetime(dataset['last_review'])

# Extraer el mes de la columna 'last_review'
dataset['month'] = dataset['last_review'].dt.month

# Contar la cantidad de revisiones por mes
reviews_by_month = dataset['month'].value_counts().sort_index()

# Configurando el estilo y la paleta de colores 
sns.set_style("whitegrid")
sns.set_palette("Set3")

# Crear el gráfico de barras o de líneas para mostrar la cantidad de revisiones por mes
plt.figure(figsize=(10, 6))
sns.lineplot(x=reviews_by_month.index, y=reviews_by_month.values, marker='o', color='tab:blue', linewidth=2, markersize=8)

# TItulo y etiquetas
plt.title('Distribución de reviews por mes', fontsize=16,fontweight='bold', pad=20, loc='left')
plt.xlabel('Mes', fontsize=14)
plt.ylabel('Cantidad de reviews', fontsize=14)
plt.xticks(range(1, 13), ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic'])

plt.tight_layout()
plt.show()
```
